```python
import numpy as N
import pandas as pd
from colorama import Fore, Back, Style
```

```python
pharm = pd.read_csv("pharm.csv")#import dataset generated by pharm_split.ipynb

pharm.head()#find the header for the diagnosis
```

```python
diagnosis = pharm['diagnosis'].tolist()#convert column, "diagnosis" to a list
```

```python
res = N.array(diagnosis) 
unique_res = N.unique(res) #count unique diagnoses in this data set
```

```python
print(*enumerate(unique_res), sep='\n')#display all unique diagnosis in this data set, with their count number
```

Below is the code to find acceptable drugs for any given diagnosis.

If there is only one acceptable drug found, the warning "There are no available replacement drugs" is printed

```python
i= 10 #i = diagnosis (0= A00.82, 1=A13.39, ... , 132=Z98.86)

drug_filter = pharm.loc[pharm['diagnosis'] == unique_res[i], 'drug']#create a filter that finds acceptable drugs for a given diagnosis

res_list = []#create an empty list

for item in drug_filter: #this for loop fills the empty list with a single entry of each unique drug for a given diagnosis
    if item not in res_list: 
        res_list.append(item) 
print("For diagnosis:", unique_res[i],"the acceptable drugs are:\n")    
for item in res_list:   
    print(item) 
length = len(res_list)    
if length < 2:
    print(Fore.RED + "\nThere are no available replacement drugs" + Style.RESET_ALL)
```

Below is the code to find acceptable drugs for all diagnoses.

If there is only one acceptable drug found, the warning "There are no available replacement drugs" is printed

```python
i=0
for i in range(0,133):
    drug_filter = pharm.loc[pharm['diagnosis'] == unique_res[i], 'drug']
    res_list = []

    for item in drug_filter: 
        if item not in res_list: 
            res_list.append(item)
    print("For diagnosis:", unique_res[i],"the acceptable drugs are:\n")
    for item in res_list: 
        print(item) 
    length = len(res_list)    
    if length < 2:
        print(Fore.RED + "\nThere are no available replacement drugs" + Style.RESET_ALL)   

    print("\n*******************\n")
    i = i+1
```
