```python
import numpy as N
import pandas as pd
from colorama import Fore, Back, Style
from collections import Counter
```

```python
import timeit
```

```python
pharm = pd.read_csv("pharm.csv")#import dataset generated by pharm_split.ipynb

pharm.head()
```

```python
diagnosis = pharm['diagnosis'].tolist()#convert column, "diagnosis" to a list
reject = pharm['patient_pay'].tolist()
```

```python
res = N.array(diagnosis) 
unique_res = N.unique(res) #count unique diagnoses in this data set
```

```python
print(*enumerate(unique_res), sep='\n')#display all unique diagnosis in this data set, with their count number
```

Create a filter which ranks approved (not rejected) drugs per given input:

```python
i= 100 #i = diagnosis (0= A00.82, 1=A13.39, ... , 132=Z98.86)

app_filter = pharm.loc[(pharm['diagnosis'] == unique_res[i]) & (pharm['patient_pay'] > 0), 'drug']#create a filter that finds acceptable drugs for a given diagnosis



print("For diagnosis:", unique_res[i],"the acceptable drugs are:\n")   
Counter(app_filter).most_common()
```

Below is the code which ranks approved drugs by the quantity of prescriptions for every diagnosis:

```python
i=0

tic=timeit.default_timer()
for i in range(0,133):
    app_filter = pharm.loc[(pharm['diagnosis'] == unique_res[i]) & (pharm['patient_pay'] > 0), 'drug']#create a filter that finds acceptable drugs for a given diagnosis

    print("For diagnosis:", unique_res[i],"the acceptable drugs are:\n")   
    print(Counter(app_filter).most_common(), sep='\n')

    print("\n*******************\n")
    i = i+1
    
toc=timeit.default_timer()
t_sec = toc - tic #elapsed time in seconds
(t_min, t_sec) = divmod(t_sec,60)
print("Process took", t_min, "minutes ", t_sec, "seconds")
```
