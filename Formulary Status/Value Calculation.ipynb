This code calculates the average patient pay for a given medication for a diagnosis under one insurance grouping(Bin, PCN, Group). This is the average a patient would be expected to pay if they don't have access to the formulary status of the drug, and are chosing a medication at random. The code then calculates the difference between the calculated average, and the price of the highest ranked formulary status, showing an average savings.

```python
ranked = pd.read_csv("Ranked.csv")
ranked.head()
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
mean = ranked.groupby(['Diagnosis','Bin','PCN','Group'])['Median Patient Pay'].mean()
mean
```

```python
min = ranked.groupby(['Diagnosis','Bin','PCN','Group'])['Median Patient Pay'].min()
min
```

```python
diff = mean.sub(min)
diff
```

The highest savings from a picking a drug treatment plan without knowing formulary status vs knowing formulary status is $241.55 for the patient's out of pocket cost.

```python
diff.sort_values(ascending=False)
```

The average savings of picking a drug treatment plan without knowing formulary status vs knowing formulary status is $20.50 for the patient's out of pocket cost.

```python
diff.mean()
```
